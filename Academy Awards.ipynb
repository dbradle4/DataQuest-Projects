```python
import pandas as pd
academy_awards = pd.read_csv("academy_awards.csv", encoding="iso-8859-1")
```

```python
academy_awards.head(5)
```

```python
print(academy_awards['Unnamed: 5'].value_counts(), 
      academy_awards['Unnamed: 6'].value_counts(),
      academy_awards['Unnamed: 7'].value_counts(),
      academy_awards['Unnamed: 8'].value_counts(),
      academy_awards['Unnamed: 9'].value_counts(),
      academy_awards['Unnamed: 10'].value_counts(),
      sep='\n')
```

```python
academy_awards['Additional Info'].value_counts()
```

```python
academy_awards['Year'] = academy_awards['Year'].str[0:4].astype('int64')
```

```python
academy_awards = academy_awards.drop(['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'], 1)
```

```python
later_than_2000 = academy_awards[academy_awards['Year'] > 2000]
```

```python
award_categories = ['Actor -- Leading Role', 'Actor -- Supporting Role', 'Actress -- Leading Role', 'Actress -- Supporting Role']
noms = later_than_2000[later_than_2000['Category'].isin(award_categories)]
```

```python
replace_dict = {'NO' : 0, 'YES': 1}
noms['Won?'] = noms['Won?'].map(replace_dict)
noms
```

```python
noms['Won'] = noms['Won?']
final_nominations = noms.drop('Won?', 1)
```

```python
additional_info_one = final_nominations['Additional Info'].str.rstrip("'}")
```

```python
additional_info_two = additional_info_one.str.split(" {'")
additional_info_two
```

```python
movie_names = additional_info_two.str[0]
characters = additional_info_two.str[1]
final_nominations['Movie'] = movie_names
final_nominations['Character'] = characters
final_nominations[['Movie', 'Character']].head(5)
```

```python
final_nominations = final_nominations.drop('Additional Info', 1)
```

```python
final_nominations.head(5)
```

```python
import sqlite3
conn = sqlite3.connect('nominations.db')
final_nominations.to_sql('nominations', conn, index=False)
```

```python
conn.execute('pragma table_info(nominations);').fetchall()
```

```python
conn.execute('select * from nominations limit 10;').fetchall()
```
